```python
import numpy
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data_file_name = 'Binance_BTCUSDT.csv'
```

```python
df = pd.read_csv(data_file_name, delimiter=';')
```

```python
df.describe()
```

```python
df.head(5)
```

```python
df['Open time tag'] = df['Open time'].apply(lambda x: pd.to_datetime(x, unit='ms'))
df['Close time tag'] = df['Close time'].apply(lambda x: pd.to_datetime(x, unit='ms'))
df[['Open time tag', 'Close time tag']]
```

```python
fig, ax = plt.subplots(2, 1, figsize=(8, 6))
ax[0].plot(df['Open time tag'], df['Open price'])
ax[0].plot(df['Open time tag'], df['Close price'])
ax[1].plot(df['Open time tag'], df['Volume'])
```
